# 🚀 Zeta Android APK Builder

Build Python app jadi APK pakai Google Colab + Buildozer + Kivy

**Estimated time:** 10-15 minutes

```
# Step 1: Install dependencies
!apt-get update
!apt-get install -y python3-dev python3-pip libg-dev libkivy-dev cmake git zip unzip
!pip install --upgrade pip
!pip install kivy buildozer
!pip install cython
```

```
# Step 2: Install Android SDK & NDK
!mkdir -p /root/android-sdk
!cd /root/android-sdk

# Download Android command line tools
!wget -q https://dl.google.com/android/repository/commandlinetools-linux-11076708_latest.zip -O cmdline-tools.zip
!unzip -q cmdline-tools.zip
!mkdir -p cmdline-tools/latest
!mv cmdline-tools/bin cmdline-tools/lib cmdline-tools/NOTICE.txt cmdline-tools/source.properties cmdline-tools/latest/ 2>/dev/null || true

# Set environment
import os
os.environ['ANDROID_HOME'] = '/root/android-sdk'
os.environ['ANDROID_SDK_ROOT'] = '/root/android-sdk'
os.environ['PATH'] = f"{os.environ['PATH']}:/root/android-sdk/cmdline-tools/latest/bin:/root/android-sdk/platform-tools"

# Accept licenses & install SDK components
!yes | /root/android-sdk/cmdline-tools/latest/bin/sdkmanager --licenses
!/root/android-sdk/cmdline-tools/latest/bin/sdkmanager "platforms;android-33" "build-tools;33.0.2" "ndk;25b"
```

```
# Step 3: Clone your project
# Replace with your repo URL
!git clone https://github.com/354ayam/zeta-android-app.git
!cd zeta-android-app && ls -la
```

```
# Step 4: Create buildozer.spec
%%writefile zeta-android-app/buildozer.spec

[app]

title = Zeta App
package.name = zetaapp
package.domain = org.zeta

source.dir = .
version = 1.0.0

requirements = python3,kivy

osx.python_version = 3
osx.kivy_version = 2.3.0

fullscreen = 1

android.permissions = INTERNET

[buildozer]
log_level = 2
warn_on_root = 1

[kdroid]
android_api = 33
ndk_api = 25
```

```
# Step 5: Build APK
!cd zeta-android-app && buildozer android debug

print("\n" + "="*50)
print("🎉 BUILD COMPLETE!")
print("="*50)
```

```
# Step 6: Download APK
from google.colab import files
import os

apk_path = 'zeta-android-app/.buildozer/android/platform/python-app/bin/'

# Find APK file
for f in os.listdir(apk_path):
    if f.endswith('.apk'):
        print(f"Found APK: {f}")
        print(f"Size: {os.path.getsize(os.path.join(apk_path, f)) / (1024*1024):.2f} MB")
        files.download(os.path.join(apk_path, f))
```
